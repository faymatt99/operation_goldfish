```python
from matplotlib import pyplot as plt;
plt.style.use('ggplot');

# Create a function which creates the relation between a categorical variable and the box plot
# As a remark, SalePrice is the target.
def category_box(df, cat="LotShape"):
    '''
    Enter the data frame and the name of a categorical variable, return a box plot between the categorical
    variable and the SalePrice.
    Input variables:
    df: the data frame. Usually just Ames_Housing_Price_Data_raw or something derived from it.
    cat: a string which is a column in df. It's the user's responsibility to make sure it is categorical.
    '''
    if cat not in df.columns:
        raise ValueError("No such column!");
    housing.boxplot(by = cat, column = "SalePrice");
    return;
```
